```python
data_3 = pd.read_csv('/content/drive/My Drive/Observing Microlensing with ATLAS/Gaia_Data/Gaia20fnr.txt')


data_3['averagemag'] = pd.to_numeric(data_3['averagemag'], errors='coerce')


df_cleaned = data_3.dropna(subset=['averagemag'])
df_cleaned.reset_index(drop=True, inplace=True)


df_cleaned
```

```python
def get_gaia_errors(mag):
    a1 = 0.2
    b1 = -5.2
    a2 = 0.26
    b2 = -6.26

    err = []
    for i in range(0, len(mag)):
        if mag[i] <= 13.5:
            err_corr = a1 * 13.5 + b1
        elif mag[i] > 13.5 and mag[i] <= 17.:
            err_corr = a1 * mag[i] + b1
        elif mag[i] > 17.:
            err_corr = a2 * mag[i] + b2
        err.append(10. ** err_corr)
    return err




df_cleaned['err_mag'] = get_gaia_errors(df_cleaned['averagemag'])
df_cleaned
```

```python
from pyLIMA.models import PSBL_model
psbl = PSBL_model.PSBLmodel(your_event)
```

```python
initial_parameters = [2459406.30627, 0.33647, 46.01927, 0.59084, 0.37058, 1.6186]
```

```python
from pyLIMA.fits import MCMC_fit
my_fit5 = MCMC_fit.MCMCfit(psbl)
my_fit5.model_parameters_guess = initial_parameters
my_fit5.fit()
```
